## 2. How many orders per month on average are not delivered for various reasons (drill down by reasons)?

We take into account orders that were paid for but never reached the buyer. That is:


- we exclude orders with "delivered" status (even those that do not have order_delivered_customer_date, as it is taken for data collection errors) and "shipped" status (these orders are still in transit, it is clear that they have not yet reached the buyer);

- from "canceled" we take only those orders that have order_approved_at (because if the order has not been paid, it has no chance to reach the buyer);

- "unavailable" are taken in full, probably these are the orders that were unavailable after payment for some reason (for example, the goods in stock ran out, but it became known after creating and paying for the order);

- "invoiced" are not taken for calculation, as these are probably orders that have not yet been received by the delivery service, i.e. they still have a chance to be delivered;

- "processing", "created" and "approved" are also not taken, because there is a chance that the order will still be delivered;

For calculating the average we take the month of order payment, because only this date is available for all selected orders.

Possible reasons for non-delivery of the order:

- order_unavailable: after payment, the order was unavailable for some reason (for example, the item ran out of stock, but this became known after the order was created and paid for);

- order_not_shipped: the order has not been transferred for delivery ("canceled" status and no order_delivered_carrier_date);

- delivery_fail: delivery service failed (status "canceled", there is order_delivered_carrier_date but no order_delivered_customer_date);

## 5. Using pandas, conduct a cohort analysis of users. 
## Between January and December, identify the cohort with the highest retention for the 3rd month. 

Cohort identifier - date of first purchase (order_approved_at)

Cohort size - month 

Reporting period - full calendar year (we should check, which one) 

Analyzed key indicator: Retention Rate

We can work with an existing table customers_orders 

```python
customers_orders.head()
```

```python
customers_orders.dtypes
```

```python
#Create a working dataframe, add a column with the month/year of purchase (we will use it to see the change 
#in Retention Rate) and a column with the cohort

cohort_analisys_df = customers_orders.filter(['customer_unique_id','order_id','order_approved_at'], axis=1)
cohort_analisys_df['month'] = customers_orders.order_approved_at.dt.to_period('M')
cohort_analisys_df['cohort'] = cohort_analisys_df.groupby('customer_unique_id')['order_approved_at'] \
                                                 .transform('min') \
                                                 .dt.to_period('M')
```

```python
cohort_analisys_df
```

```python
#Let's check which full year can be taken for the analysis
cohort_analisys_df.month.sort_values().unique()
```

```python
#As we can see from the data, we can work with 2017. Let's select only purchases and cohorts for 2017
cohort_analisys_df = cohort_analisys_df.query("'2016-12' < month < '2018-01' & '2016-12' < cohort < '2018-01'")
cohort_analisys_df
```

```python
#We group the data by cohort and month and calculate, how many customers are in each group
cohort_analisys = cohort_analisys_df.groupby(['cohort', 'month'], as_index = False) \
                                    .agg({'customer_unique_id': 'count'}) \
                                    .rename(columns = {'customer_unique_id': 'customers'})

#We add a column, in which we put the number of months, that have passed from the first purchase, for each group
from operator import attrgetter
cohort_analisys['period_number'] = (cohort_analisys.month - cohort_analisys.cohort).apply(attrgetter('n'))
```

```python
cohort_analisys
```

```python
#Let's create a pivot to get the final table with Retention Rate
cohort_analisys_pivot = cohort_analisys.pivot_table(index='cohort',
                                     columns='period_number',
                                     values='customers')
```

```python
cohort_analisys_pivot
```

```python
#Let's calculate Retention Rate in % of the zero month
cohort_size = cohort_analisys_pivot.iloc[:, 0]
retention_matrix = cohort_analisys_pivot.divide(cohort_size, axis=0)*100
```

```python
retention_matrix
```

As can be seen from the table above, the cohort with the highest retention at month 3 is '2017-09'

## Using python, build RFM segmentation of users to qualitatively evaluate your audience. In clustering, you can select the following metrics: R - time from the user's last purchase to the current date, F - total number of purchases from the user over time, M - sum of purchases over time. Describe in detail how you created the clusters. For each RFM segment, build the boundaries of recency, frequency and monetary metrics to interpret these clusters. 

First we need to merge all three tables: customers, orders, items

```python
#Now, let's select the necessary data from the table customers_orders
customers_orders_rfm_data = customers_orders.filter(['customer_unique_id','order_id','order_approved_at'], axis=1)
```

```python
#Then we aggregate the data from the items table to get the sum of purchases for each order to calculate M
items_rfm_data = items.groupby(['order_id', 'product_id'], as_index = False).agg({'price': 'sum'}).rename(columns = {'price': 'product_sum'})
```

```python
items_rfm_data = items_rfm_data.groupby('order_id', as_index = False).agg({'product_sum':'sum'}).rename(columns = {'product_sum':'order_sum'})
```

```python
items_rfm_data
```

```python
#We merge the dataframes. The tables have different numbers of unique order_ids, we will keep only those, 
#for which there is data in both tables
```

```python
rfm_data = customers_orders_rfm_data.merge(items_rfm_data, on = 'order_id', how = 'inner')
rfm_data
```

```python
#Let's check the period for which data are available
print('Orders from {} to {}'.format(rfm_data['order_approved_at'].min(),
                                    rfm_data['order_approved_at'].max()))
```

```python
#For the calculation, let's take 365 days till 2018-09-03 as standard RFM analysis is done for the last year
period = 365
```

```python
#To calculate recency, let's set the date, from which we will count, as the day following the last day of 
#the last payment in our data
NOW = rfm_data.order_approved_at.max() + timedelta(days=1)
NOW
```

Calculate Recency as the number of days since last purchase

```python
#To do this, let's add a column to our table with the number of days between each order and NOW
rfm_data['days_since_order'] = rfm_data.order_approved_at.apply(lambda x: (NOW - x).days)
rfm_data
```

```python
#We filter out customers who have shopped in our period
rfm_data = rfm_data.loc[rfm_data['order_approved_at'] >= NOW - timedelta(days=period)]
```

```python
rfm_data
```

```python
#Now let's find the minimum days_since_order value for each customer (their last purchase). This will be the Recency value
#Also calculate the total number of orders for the last 365 days (period) for each customer. This will be the Frequency value
aggr = {
    'days_since_order': lambda x: x.min(),
    'order_approved_at': lambda x: len([d for d in x if d >= NOW - timedelta(days=period)]),
}
rfm = rfm_data.groupby('customer_unique_id').agg(aggr).reset_index()
rfm.rename(columns={'days_since_order': 'Recency', 'order_approved_at': 'Frequency'}, inplace=True)
rfm.head()
```

```python
#Calculate Monetary as the sum of the values of all orders each customer has placed in the last 365 days
rfm['Monetary'] = rfm['customer_unique_id'].apply(lambda x: rfm_data[(rfm_data['customer_unique_id'] == x) & \
                                                           (rfm_data['order_approved_at'] >= NOW - timedelta(days=period))]\
                                                           ['order_sum'].sum())
rfm.head()
```

Monetary distribution looks more like Frequency rather than Recency. Judging by the graph, most of the values are in the range from 0 to 250 (apparently, it's reals, not rubles, judging by the geography of the data :).

Let's single out the following groups:

1)0-250

2)251-1750

3)1751 and more

```python
rfm[rfm.Monetary > 1750]
```

```python
#Let's write functions to further attribute customers to the groups
def r_score(x):
    if x <= third_r[.33]:
        return 1
    elif x <= third_r[.66]:
        return 2
    else:
        return 3
    
def f_score(x):
    if x == 1:
        return 3
    elif x == 2:
        return 2
    else:
        return 1
    
def m_score(x):
    if x <= 250:
        return 3
    elif x <= 1750:
        return 2
    else:
        return 1
```

```python
#Применим функции к нашим данным
rfm['R'] = rfm.Recency.apply(lambda x: r_score(x))
rfm['F'] = rfm.Frequency.apply(lambda x: f_score(x))
rfm['M'] = rfm.Monetary.apply(lambda x: m_score(x))
```

```python
rfm
```

```python
#Finally, we get the segment number for each customer
rfm['RFM_score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
rfm.head()
```
